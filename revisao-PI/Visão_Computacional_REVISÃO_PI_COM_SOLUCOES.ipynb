**PROBLEMA 2.** Considere a função abaixo, que recebe quatro parâmetros: uma imagem em níveis de cinza, a coordenada de um pixel (x,y) e um valor T.

```
def enigma(img,x,y,T):
  if abs(img[x+1][y]-img[x][y])>=T:
    return 255
  else:
    return 0
```

Utilize esta função para calcular um contorno da imagem lena.png, lida no código abaixo:

```
import cv2 as cv
from google.colab.patches import cv2_imshow
img = cv.imread('lena.png')
print(img.shape)
cv2_imshow(img)
```

```
import numpy as np
img = cv.cvtColor(cv.imread('lena.png'), cv.COLOR_BGR2GRAY).astype(np.int16)
rows,columns=img.shape
for x in range(rows-1):
  for y in range(columns):
    img[x][y]=enigma(img,x,y,20)
cv2_imshow(img)
```

**PROBLEMA 4.** Abaixo, temos um código em OpenCV que abre uma imagem RGB com três canais de cor, cada cor representada como um número inteiro. Escreva um programa em OpenCV que transforme esta imagem para níveis de cinza, manipulando diretamente os pixels da imagem e sem usar as funções de conversão de cor da biblioteca.

```
import cv2 as cv
from google.colab.patches import cv2_imshow
img = cv.imread('lena.png')
print(img.shape)
cv2_imshow(img)
```

(R,G,B) ou (B,G,R) 
(C,C,C) cinza (0,0,0) - preto   (255,255,255) - branco

d=sqrt((B-C)^2+(G-C)ˆ2+(R-C)ˆ2)  encontrar C que minimize d

```
import math

def gray(b,g,r):
  min=math.sqrt((b-0)**2+(g-0)**2+(r-0)**2)
  gmin=0
  for gl in range (1,255):
    if math.sqrt((b-gl)**2+(g-gl)**2+(r-gl)**2)<min:
      min=math.sqrt((b-gl)**2+(g-gl)**2+(r-gl)**2)
      gmin=gl
  return [gmin,gmin,gmin]

b,g,r= cv.split(img)
rows,columns,channels=img.shape
for l in range(240,300):
  for c in range(300,360):
    img[l][c]=gray(b[l][c],g[l][c],r[l][c])
cv2_imshow(img)
```
